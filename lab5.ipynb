# Задача 1 
Разделить таблицу данных так, чтобы в обучающую выборку попало 66% наблюдений, а в тестовую выборку 33% наблюдений


```python
import random
import pandas as pd
import numpy as np

df = pd.read_csv('/Users/d.kapanadze/Downloads/Учеба/ИТМО/1 семестр/Анализ данных на Python/lectures/lect_02/1_beverage/beverage_r.csv', sep=';')

df.head()
```

```python
# фиксируем зерно
np.random.seed(seed=10)

# массив размера len(df) с выборкой из равномерного распределения
r = np.random.uniform(0, 1, size=len(df))

# Сравниваем значения с квантилем уровня 66
q66 = [i <= np.quantile(r, 0.66) for i in r]
```

```python
train = df[q66]
print(f"Размер обучающей выборки: {round(train.shape[0] / df.shape[0], 3)}")

test = df[np.logical_not(q66)]
print(f"Размер тестовой выборки: {round(test.shape[0] / df.shape[0], 3)}")
```

# Задача 2
Разделить таблицу данных на четыре одинаковых части.


```python
r2 = np.random.uniform(0, 1, size=len(df))

q25 = [ i <= np.quantile(r2, 0.25) for i in r2]
q50 = [ i <= np.quantile(r2, 0.50) and i > np.quantile(r2, 0.25) for i in r2]
q75 = [ i <= np.quantile(r2, 0.75) and i > np.quantile(r2, 0.50) for i in r2]
q100 = [ i <= np.quantile(r2, 1) and i > np.quantile(r2, 0.75) for i in r2]
```

```python
part1 = df[q25]
part2 = df[q50]
part3 = df[q75]
part4 = df[q100]

print(f"Размер 1й части: {round(part1.shape[0]/df.shape[0], 3)}")
print(f"Размер 2й части: {round(part2.shape[0]/df.shape[0], 3)}")
print(f"Размер 3й части: {round(part3.shape[0]/df.shape[0], 3)}")
print(f"Размер 4й части: {round(part4.shape[0]/df.shape[0], 3)}")
```

# Задача 3
Разделить таблицу данных на две стратифицированные подвыборки.
По переменным кока-кола и минералка.

```python
df.head()
```

```python
# Какой процент любителей кока-колы в датасете
print(f'Кола: {round(df.query("COKE == 1").shape[0] / df.shape[0], 3)}')

# Какой процент любителей минералки в датасете
print(f'Минералка: {round(df.query("TAB == 1").shape[0] / df.shape[0], 3)} ')
```

Кока-Кола(обучающая выборка - 60%, тестовая - 40%)

```python
sz1 = np.random.uniform(0, 1, size=len(df.query("COKE == 1")))
sz0 = np.random.uniform(0, 1, size=len(df.query("COKE == 0")))


train_coke = pd.concat(
    [
    df.query("COKE == 1")[sz1 < np.quantile(sz1, 0.6)],
    df.query("COKE == 0")[sz0 < np.quantile(sz0, 0.6)]
    ],
    axis=0
)

assert np.isclose(train_coke.shape[0] / df.shape[0], 0.6, 0.05)
assert np.isclose(train_coke.query("COKE == 1").shape[0] / train_coke.shape[0], df.query("COKE == 1").shape[0] / df.shape[0], 0.05)
```

```python
test_coke = pd.concat(
    [
    df.query("COKE == 1")[sz1 < np.quantile(sz1, 0.4)],
    df.query("COKE == 0")[sz0 < np.quantile(sz0, 0.4)]
    ],
    axis=0
)

assert np.isclose(test_coke.shape[0] / df.shape[0], 0.4, 0.05)
assert np.isclose(test_coke.query("COKE == 1").shape[0] / test_coke.shape[0], df.query("COKE == 1").shape[0] / df.shape[0], 0.05)
```

Минералка(обучающая выборка - 60%, тестовая - 40%)

```python
sz3 = np.random.uniform(0, 1, size=len(df.query("TAB == 1")))
sz4 = np.random.uniform(0, 1, size=len(df.query("TAB == 0")))


train_tab = pd.concat(
    [
    df.query("TAB == 1")[sz3 < np.quantile(sz3, 0.6)],
    df.query("TAB == 0")[sz4 < np.quantile(sz4, 0.6)]
    ],
    axis=0
)

assert np.isclose(train_tab.shape[0] / df.shape[0], 0.6, 0.05)
assert np.isclose(train_tab.query("TAB == 1").shape[0] / train_tab.shape[0], df.query("TAB == 1").shape[0] / df.shape[0], 0.08)
```

```python
test_tab = pd.concat(
    [
    df.query("TAB == 1")[sz3 < np.quantile(sz3, 0.4)],
    df.query("TAB == 0")[sz4 < np.quantile(sz4, 0.4)]
    ],
    axis=0
)

assert np.isclose(test_tab.shape[0] / df.shape[0], 0.4, 0.05)
assert np.isclose(test_tab.query("TAB == 1").shape[0] / test_tab.shape[0], df.query("TAB == 1").shape[0] / df.shape[0], 0.05)
```
